```python
#This gives the Azimuth and Altitude of a satellite at a specific time. What is needed is the name of the satellite, 
#the first and second lines of the TLE set, and the exact time to the second. The terms Hb, Yg, Ke, Cd and Bt then need
#to be set to either 1 or 0 depending on whether one desires output from those locations.

def VisibilityAtTime(ObjectName,line1,line2,Hb,Yg,Ke,Cd,Bt,year, month, day, hour, minute, second):

    import skyfield
    from skyfield.api import load, wgs84, EarthSatellite
    import numpy as np
    from datetime import datetime
    from pytz import timezone
    #mountpleasant = wgs84.latlon(-42.8055, 147.438)
    #newZealand = wgs84.latlon(-45.03861, 170.0958)
    #Yarragadee = wgs84.latlon(-29.04567, 115.34876)
    #Katherine = wgs84.latlon(-14.375464, 132.152373)
    #Ceduna = wgs84.latlon(-31.86883, 133.809799)
    #BisdeeTier = wgs84.latlon(-42.4311,147.2878)
    #mountpleasant = wgs84.latlon(-42.8055, 147.438)
    mountpleasant = wgs84.latlon(latitude_degrees = -42.8055808638581, longitude_degrees = 147.43813722942386, elevation_m = 40.983)
    #newZealand = wgs84.latlon(-45.03861, 170.0958)
    
    hobart = wgs84.latlon(latitude_degrees = -42.8055808638581, longitude_degrees = 147.43813722942386, elevation_m = 40.983)
    Katherine = wgs84.latlon(latitude_degrees=-14.3766210062618, longitude_degrees=132.15431778026138, elevation_m=188.978)
    Yarragadee = wgs84.latlon(latitude_degrees=-29.04713995970614, longitude_degrees=115.3456309353691, elevation_m=248.23692528494942)
    Ceduna = wgs84.latlon(latitude_degrees=-31.86767880595407, longitude_degrees=133.80983868862833, elevation_m=164.6178250238288)
    #BisdeeTier = wgs84.latlon(latitude_degrees=-42.43, longitude_degrees=147.29, elevation_m=646.0)
    BisdeeTier = wgs84.latlon(latitude_degrees=-42.4311, longitude_degrees=147.2878, elevation_m=646.0)
    
    #Yarragadee = wgs84.latlon(-29.04567, 115.34876)
    #Katherine = wgs84.latlon(-14.375464, 132.152373)
    #Ceduna = wgs84.latlon(-31.86883, 133.809799)
    #BisdeeTier = wgs84.latlon(-42.4311,147.2878)
    
    ts = load.timescale()

    satellite = EarthSatellite(line1, line2, ObjectName, ts)

    differenceHb = satellite - mountpleasant
    differenceNZ = satellite - newZealand
    differenceYg = satellite - Yarragadee
    differenceKe = satellite - Katherine
    differenceCd = satellite - Ceduna
    differenceBt = satellite - BisdeeTier
    
    observationtime = ts.utc(year, month, day, hour, minute, second)
    days = observationtime - satellite.epoch
    
    print(satellite)
    print('Time is {:.3f} days after TLE epoch time'.format(days))
    
    topocentricHb = differenceHb.at(observationtime)
    altHb, azHb, heightHb = topocentricHb.altaz()
    topocentricNZ = differenceNZ.at(observationtime)
    altNZ, azNZ, heightNZ = topocentricNZ.altaz()
    topocentricYg = differenceYg.at(observationtime)
    altYg, azYg, heightYg = topocentricYg.altaz()
    topocentricKe = differenceKe.at(observationtime)
    altKe, azKe, heightKe = topocentricKe.altaz()
    topocentricCd = differenceCd.at(observationtime)
    altCd, azCd, heightCd = topocentricCd.altaz()
    topocentricBt = differenceBt.at(observationtime)
    altBt, azBt, heightBt = topocentricBt.altaz()
    if Hb == 1:
        print("At the Hobart 12m Telescope,", "observation date:", day, month, year, "observation time", hour,":", minute,":", second,)
        print("|Altitude = ", altHb.degrees, "degrees.| Azimuth = ", azHb.degrees, "degrees.|" )
    if Yg == 1:
        print("At the Yarragadee 12m Telescope,", "observation date:", day, month, year, "observation time", hour,":", minute,":", second,)
        print("|Altitude = ", altYg.degrees, "degrees.| Azimuth = ", azYg.degrees, "degrees.|" )
    if Ke == 1:
        print("At the Katherine 12m Telescope,", "observation date:", day, month, year, "observation time", hour,":", minute,":", second,)
        print("|Altitude = ", altKe.degrees, "degrees.| Azimuth = ", azKe.degrees, "degrees.|" )
    if Cd == 1:
        print("At the Ceduna 30m Telescope,", "observation date:", day, month, year, "observation time", hour,":", minute,":", second,)
        print("|Altitude = ", altCd.degrees, "degrees.| Azimuth = ", azCd.degrees, "degrees.|" )
    if Bt == 1:
        print("At the Bisdee Tier 7.3m Telescope,", "observation date:", day, month, year, "observation time", hour,":", minute,":", second,)
        print("|Altitude = ", altBt.degrees, "degrees.| Azimuth = ", azBt.degrees, "degrees.|" )
```
